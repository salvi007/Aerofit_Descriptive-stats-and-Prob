```python
#Q4 Representing the marginal probability like - what percent of customers have purchased KP281, KP481, or KP781 in a table
pd.crosstab(df["Gender"],df["Product"],margins=True,normalize=True)  #Marginal and Joint Prob
```

**Marginal Probability**

p(Male): 0.58

p(Female) : 0.42

**Joint Probability**

p(Female & KP281) = 0.2

p(Female & KP481) = 0.1

p(Female & KP281) = 0.03

p(Male & KP281) = 0.2

p(Male & KP481) = 0.17

p(Male & KP781) = 0.18

```python
pd.crosstab(df["Gender"],df["Product"],margins=True,normalize='index')  #conditional Prob
```

**Conditional Probability**

p(Female | KP281) =0.53

p(Female | KP481)= 0.39

P(Female | KP781) = 0.09

p(Male | KP281) = 0.39

p(Male | KP481) = 0.30
